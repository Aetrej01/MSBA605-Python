##### Prompt 6

Nest an if/elif/else loop within your for loop from Prompt5 that will print 'This movie is funny' if the MAIN_GENRE is comedy, 'This movie is scary' if the MAIN_GENRE is horror OR thriller, and will pass if the MAIN_GENRE is any other value.

```python
for main_genre in movies['MAIN_GENRE']:
    if main_genre =='comedy':
        print('This movie is funny')
    elif main_genre in ['horror','thriller']:
        print('This movie is scary')
    
```

##### Prompt 9
