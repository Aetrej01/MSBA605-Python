##### Prompt3
Set your random number generator seed to 8563 and create a new variable called rarray2 with a 1 by 3 array with possible numbers from 0 to 100. Print this array.

```python
np.random.seed(8563)
rarray2=np.random.randint(low=0, high=100, size=(1,3))
print(rarray2)
```

##### Prompt4
Use broadcasting to divide rarray by rarray2 and save the result as a variable called resarray. Print this array.

```python
resarray=(rarray/rarray2)
print('divide rarray by rarray2: \n')
print(resarray)
```

##### Prompt6
Use for loops to iterate through the rows and columns of resarray and print each element. My solution (there may be others) used two for loops, a combination of range() and len(), and slicing to accomplish this task.

```python
for value in range(0,len(resarray)):
    for value2 in range(0,len(resarray[value])):
        print(resarray[value][value2])

```

##### Prompt7
Calculate the sum of the rows of resarray and save it in a variable called sumarray. Print this array.

```python
sumarray=(resarray.sum(axis=1))
print('the sum of the rows is: ')
print(sumarray)
```

##### Prompt8
Broadcast sumarray over resarray to subtract each row sum from each element in that row. You will have to change the shape of sumarray, I used .reshape().

```python
b_sum = sumarray.reshape(-1, 1)
resarray -= b_sum
print(resarray)
```

##### Prompt9
Set your random number generator seed to 7245 and create a new variable called rarray3 with a 2 by 2 array with possible numbers from 0 to 100. Print this array. Then use the linalg library to get the inverse of this array. Print the result.

```python
np.random.seed(7245)
rarray3=np.random.randint(low=0, high=100, size=(2,2))
print('new variable: \n',rarray3,'\n')
print('the inverse of rarray3: \n', np.linalg.inv(rarray3))
```

##### Prompt10
Use np.where() to create a boolean array with 1's when the elements of resarray are greater than or equal to the overall mean of resarray and 0's when they are less than.

```python
resarray=(rarray/rarray2)
mean_R=np.mean(resarray)
print('The mean is: \n', mean_R )
bool_array=np.where(resarray>=mean_R,1,0)
print('The boolean array is: \n', bool_array)
```
