##### Prompt 3
What genres of movies are present in both the 20th century and 21st century movies?

```python
#pelis.query("`Century` == 'coca cola' or `product_name` == 'pepsi'")
```

```python
th_century = set(pelis[pelis['Century'] == '20th Century']['MAIN_GENRE'].unique())
st_century = set(pelis[pelis['Century'] == '21st Century']['MAIN_GENRE'].unique())

century_genres = th_century.intersection(st_century)

pd.DataFrame(century_genres)
```

##### Prompt 8

Merge the DataFrame with the age certification with our Movie DataFrame so that only Titles present in the Movie DataFrame are present in the resulting DataFrame.

##### Prompt 10

By happenstance (e.g., this will not always be the case), it appears that the television version of ratings are all the last entry for that title. Drop duplicates keeping the first value of each title.

Then Group the titles by age_certification to determine which age_certification has the longest average duration.

```python
merged_pelis=merged_pelis.drop_duplicates(['TITLE'], keep='first')
merged_pelis
```

```python

certificate=merged_pelis.groupby('age_certification')['DURATION'].mean()
print('Age certification with longest average:', certificate.idxmax(), certificate.max())
```
