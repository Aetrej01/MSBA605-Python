#### Prompt3
Write a function called `oddoreven` using an `if`/`else` statement to determine if the number passed in the function is even or odd. Make sure the function prints `'This is an even number.'` if the number is even and `'This is an odd number.'` if the number is odd. Test your function with both an odd and an even number to ensure it works. (**Note:** The modulus operator `%` will return the remainder of a division problem; for even numbers, the result is 0.)

```python
def oddoreven(num: int):
    if num %2 !=0:
        print(f'This is an odd number: {num}.')
    elif num %2 ==0:
        print(f'This is an even number: {num}.')

oddoreven(4)
oddoreven(7)
```

####Prompt7

Create two arrays. One called students with `'Janet'`, `'Adriana'`, `'Manual'`, `'Mohamed'`, `'Leann'`, `'Samuel'` as values. Another called grades with `[93, 85]`, `[78, 80]`, `[94, 93]`, `[75, 90]`, `[92, 87]`, `[96, 87]` as values.

Select all rows from `grades` where `students` is either `'Janet'` or `'Samuel'`

```python
students=np.array(['Janet', 'Adriana', 'Manual', 'Mohamed', 'Leann', 'Samuel'])
grades=np.array([[93, 85], [78, 80], [94, 93], [75, 90], [92, 87], [96, 87]])

arrayG=np.isin(students, ['Janet', 'Samuel'])
Select=grades[arrayG]
print(Select)
```

### Prompt8

Use `np.where()` to create a Boolean array that will print `True` when the student's grade is greater than the overall class average (or `mean`) and `False` when it is not. Print both the overall class average and your Boolean array.

```python
mean_G=np.mean(grades)
print('The mean is: \n', mean_G)
bool_G=np.where(grades>mean_G, 'True','False')
print('The boolean grade is:\n', bool_G)
```
