##### Prompt 2

There is a column called 'data_quality' which indicates which of our rows contains data that is reliable.

Explore this variable a little by: determining the number of unreliable vs. reliable rows, the country (or countries) with the highest number of unreliable rows, then the country (or countries) with the highest proportion of unreliable rows (# of unreliable rows/# of total rows for each country).

We do this to understand a little more able how reliable our data is, and what some of the consequence will be if we choose to remove unreliable data.


```python
#Determining the number of unreliable vs. reliable rows
rel_r = cost_of_living[cost_of_living['data_quality'] == 1].shape[0]
unrel_r = cost_of_living[cost_of_living['data_quality'] == 0].shape[0]
print(f"Number of reliable rows: {rel_r}")
print(f"Number of unreliable rows: {unrel_r}")

#The country (or countries) with the highest number of unreliable rows
unrel_country = cost_of_living[cost_of_living['data_quality'] == 0]['Country'].value_counts()
highest_unrel = unrel_country.idxmax()
print(f"Country with the highest number of unreliable rows: {highest_unrel}")

#The country (or countries) with the highest proportion of unreliable rows (# of unreliable rows/# of total rows for each country)
total_r = cost_of_living['Country'].value_counts()
unreliable = (unrel_country / total_r).dropna()
country_unreliable = unreliable.idxmax()
print(f"Country with the highest proportion of unreliable rows: {country_unreliable}")
```

##### Prompt 3

Now that we know a little more about our data in terms of reliability, we will want to create a new DataFrame that contains only rows where 'data_quality' equals 1 (e.g., data is reliable). Then drop the data_quality column. This way we can use only reliable data in our subsequent analyses.

Make sure to print the first 5 rows of your new DataFrame by using the .head() command and use this DataFrame in all subsequent prompts.

```python
reliable_r = cost_of_living[cost_of_living['data_quality'] == 1]
reliable_r = reliable_r.drop(columns=['data_quality'])#I'm deleting the data_quality cell, .drop means deleting the column
reliable_r.head()
```

Create a new column called 'City/Country' that combines City and Country with a comman in between.

```python
reliable_r['City/Country'] =reliable_r['City']+', '+reliable_r['Country']
reliable_r.head() #or reliable_r in case I want to see all rows
```

##### Prompt 5

In how many cities is it cheaper to buy beef than it is to buy chicken?

```python
cheaper_buy_beef = reliable_r[reliable_r['Beef'] < reliable_r['Chicken']]['City'].nunique() #nunique returns the number of unique elements in a given series or dataframe
print(f"In {cheaper_buy_beef} cities, it is cheaper to buy beef than it is to buy chicken.")
```

##### Prompt 7

Create two new columns: One called Avg_Fruit that contains the average of Apples, Bananas, and Oranges and another called Avg_Veg that contains the average of 'Tomato', 'Potato', 'Onion', 'Lettuce'.

Note: Depending on your method, you may need to use axis=1.

```python
reliable_r['Avg_Fruit'] = reliable_r[['Apples', 'Bananas', 'Oranges']].mean(axis=1).round(2)
reliable_r['Avg_Veg'] = reliable_r[['Tomato', 'Potato', 'Onion', 'Lettuce']].mean(axis=1).round(2)
reliable_r.head()
```

##### Prompt 8

```python
highest_avgfruit = reliable_r.groupby('Country')['Avg_Fruit'].mean().idxmax()
print(f"{highest_avgfruit} has the highest average fruit cost.")
```

##### Prompt 9
The cost of living data set we have been using was updated. While we could continue to keep the DataFrames separate, we want to concatenate them so that all data is contained in one DataFrame.

To prepare for this merge, first add a column called 'year' with values of 2021 to our current DataFrame.

Next, read in the file called 'CostofLivingforWeek3_2.csv'. Add a column called 'year' with values of 2022 to your import and remove any rows were data_quality is equal to 0.

Finally, concatenate your DataFrames so the new data appears as additional rows.

```python
reliable_r['year'] = 2021
reliable_r.head()
```

```python
cost2 = pd.read_csv('CostofLivingforWeek3_2.csv')
cost2['year'] = 2022
cost2 = cost2[cost2['data_quality'] == 1]
cost2.head()
```

##### Prompt 10

Determine if the average cost of fruit has changed from 2021 to 2022.

```python
fruit2021 = reliable_r[reliable_r['year'] == 2021]['Avg_Fruit'].mean()
fruit2022 = reliable_r[reliable_r['year'] == 2022]['Avg_Fruit'].mean()

if fruit2021 == fruit2022:
    print("The average cost of fruit has not changed from 2021 to 2022.")
elif fruit2021 < fruit2022:
    print("The average cost of fruit has increased from 2021 to 2022.")
else:
    print("The average cost of fruit has decreased from 2021 to 2022.")
```
