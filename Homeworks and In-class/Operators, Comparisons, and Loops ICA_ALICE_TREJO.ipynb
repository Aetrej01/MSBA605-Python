##### Prompt3
Now create a list called lis1 with the values 4, 5, 7, 8, 9, 10. Write a for loop that prints whether each element in the list is greater than or equal to var2.

```python
lis1 = [4,5,7,8,9,10]

for value in lis1:
    if value >=  var2:
        print(value)

```

##### Prompt4
Now within the for loop you created above, add an if/else statement that prints 'This number is greater' if the number is greater than var2, and 'This number is less' if the number is less than var2.

```python
for value in lis1:
    if value > var2:
        print(value, 'is greater than', var2)
    elif value < var2:
        print(value, 'is less than', var2)
```

##### Prompt5
You may have realized that the above output is not quite correct. We need to add an elif statement to capture when the numbers are equal. Adjust the above loop to be more accurate by changing the if to greater than and adding an elif testing if the numbers of equal and printing 'The numbers are equal' if it is true.

```python
for value in lis1:
    if value > var2:
        print(value, 'is greater than', var2)
    elif value < var2:
        print(value, 'is less than', var2)
    elif value == var2:
        print(value, "are equal", var2)
```
