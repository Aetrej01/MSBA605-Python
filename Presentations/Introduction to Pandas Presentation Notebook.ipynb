#### Series
- Data structure with an index and one column

- Aligns on index
- Useful since when accessing a column from a DataFrame, it acts as a Series


```python
# Defining Series
s = pd.Series([1,2,3,4], index=['a','b','c','d'])
print(s)
```

```python
# Accessing data from Series
print(s[0]) #or print(s['a'])
```

```python
#Math with series - aligns on index
s2=pd.Series([10,20,30,40], index=['a','b','d','c'])
print(s2)
```

```python
print(s+s2)
```

#### DataFrame
- mimics tabular data in programming

- Can be constructed but files are read in automatically as DataFrames


```python
#Constructing dataframe using pd.DataFrame
df2=pd.DataFrame(np.array([('14','4'),('Chorkie','Patterdale Terrier')]),columns=['Delilah','Kemba'],index=['age','breed'])
print(df2)
```

```python
#Reading in file as DataFrame
degdate=pd.read_excel('DegreeDates.xlsx')
print(degdate)
```

#### Index Manipulation
- Reindex
    - will coerce missing value
- Set index
    - uses existing column in DataFrame

```python
# My Code
# .reindex()
df2=df2.reindex(['age','breed','favtreat'])
print(df2)
```

```python
# .set_index()
degdate1=degdate.set_index('ID',drop=True)
print(degdate1)
```

```python
# Practice
```

#### Dropping and Slicing

```python
# My Code
# Drop
    # row
df3=df2.drop('favtreat')
print(df3)
```

```python
    # column
df4=df3.drop(columns='Kemba')
print(df4)
```

```python
# Slicing
    # columns
print(df2['Delilah'])
```

```python
    # index
print(df2[1:3])
```

```python
    # conditional
print(degdate1[degdate1['GRADTERM']=='MAY 17'])
```

```python
# Practice
```

#### Sorting and Rank

```python
#Sort
    #by index
degdate2=degdate1.sort_index(ascending=False)
print(degdate2)
```

```python
    #by column
degdate2=degdate1.sort_values(by='GRADTERM')
print(degdate2)
```

```python
#Rank
degdate['ID rank']=degdate['ID'].rank()
print(degdate)
```

```python
#Practice
```

#### Functions
- you can apply functions across the entire DataFrame or specific columns
- there are also built in functions

```python
#Custom
# for numeric functions, you must limit to numeric types
degdate2=degdate[['ID','ID rank']]
```

```python
# .apply()
print(degdate2.apply(lambda x: x.max()-x.min()))
```

```python
# .applymap()
print(degdate2.applymap(lambda x: x*10))
```

```python
# .map()
print(degdate2['ID rank'].map(lambda x: x*10))
```

```python
#Built in
#string
    # isin()
print(degdate['GRADTERM'].isin(['MAY 17','MAY 18','MAY 19','MAY 20']), '\n')
```

```python
    #value_counts()
print(degdate['GRADTERM'].value_counts(), '\n')
```

```python
# mathmatical
    #log
print(np.log(degdate2), '\n')
```

```python
    #sum
print(degdate2.sum(), '\n')
```

```python
    #mean
print(degdate2.mean(), '\n')
```

```python
    #describe
print(degdate2.describe(), '\n')
```
