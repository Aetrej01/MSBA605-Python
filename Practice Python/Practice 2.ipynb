```python
import pandas as pd


def process_response_dict(response_dict: dict) -> pd.DataFrame:
        columns_dict = {
                "concert_id": [],
                "concert_title": [],
                "concert_popularity": [],
                "concert_datime_utc": [],
                "performer_id": [],
                "performer_name": [],
                "performer_score": [],
                "performer_popularity": [],
                "venue_name": [],
                "venue_city": [],
                "venue_state": [],
                "venue_postal_code": [],
                "venue_address": [],
                "listing_count": [],
                "average_price": [],
                "lowest_price_good_deals": [],
                "lowest_price": [],
                "highest_price": [],
                "visible_listing_count": [],
                "median_price": [],
                "lowest_sg_base_price": [],
                "lowest_sg_base_price_good_deals": [],
                "concert_score": [],
                "announce_date": [],
                "created_at": [],
        }

        for event in response_dict["events"]:
                columns_dict["concert_id"].append(event["id"])
                columns_dict["concert_title"].append(event["title"])
                columns_dict["concert_popularity"].append(event["popularity"])
                columns_dict["concert_datime_utc"].append(event["datetime_utc"])
                for performer in event["performers"]:
                        if performer.get("primary"):
                                columns_dict["performer_id"].append(performer["id"])
                                columns_dict["performer_name"].append(performer["name"])
                                columns_dict["performer_score"].append(performer["score"])
                                columns_dict["performer_popularity"].append(performer["popularity"])
                                
                columns_dict["venue_name"].append(event["venue"]["name"])
                columns_dict["venue_city"].append(event["venue"]["city"])
                columns_dict["venue_state"].append(event["venue"]["state"])
                columns_dict["venue_postal_code"].append(event["venue"]["postal_code"])
                columns_dict["venue_address"].append(event["venue"]["address"])
                columns_dict["listing_count"].append(event["stats"]["listing_count"])
                columns_dict["average_price"].append(event["stats"]["average_price"])
                columns_dict["lowest_price_good_deals"].append(event["stats"]["lowest_price_good_deals"])
                columns_dict["lowest_price"].append(event["stats"]["lowest_price"])
                columns_dict["highest_price"].append(event["stats"]["highest_price"])
                columns_dict["visible_listing_count"].append(event["stats"]["visible_listing_count"])
                columns_dict["median_price"].append(event["stats"]["median_price"])
                columns_dict["lowest_sg_base_price"].append(event["stats"]["lowest_sg_base_price"])
                columns_dict["lowest_sg_base_price_good_deals"].append(event["stats"]["lowest_sg_base_price_good_deals"])
                columns_dict["concert_score"].append(event["score"])
                columns_dict["announce_date"].append(event["announce_date"])
                columns_dict["created_at"].append(event["created_at"])

        return pd.DataFrame(columns_dict)
        pd.DataFrame(columns_dict).to_csv("results.csv", index=False)
```
